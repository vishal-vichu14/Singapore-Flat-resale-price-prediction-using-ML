```python
# Function to calculate remaining lease based on year and lease commencement date
def remaining(data):
    data['year'] = data['month'].str[:4]
    data['year'] = data['year'].astype(int)
    data['remaining_lease'] = 99 - (data['year'] - data['lease_commence_date'])
    data = data.drop(['year'], axis = 1)
    return data
```

```python
# Apply remaining lease calculation to the first three datasets
data1 = remaining(data1)
data2 = remaining(data2)
data3 = remaining(data3)
```

```python
# Function to calculate the median of the storey range (e.g., '5 TO 10' -> (5+10)/2)
def calculate_mean(range_str):
    start, end = range_str.split(' TO ')
    return (int(start) + int(end)) / 2 
```

```python
# Apply the storey range median calculation
combined_df['storey_mean'] = combined_df['storey_range'].apply(calculate_mean)
```

```python
import requests
import json

mrt_lat = [] # To store MRT station latitudes
mrt_long = [] # To store MRT station longitudes

# Loop through each MRT station to get its latitude and longitude using the OneMap API
for query_address in list_of_mrt:
    query_string = 'https://www.onemap.gov.sg/api/common/elastic/search?searchVal='+str(query_address)+'&returnGeom=Y&getAddrDetails=Y&pageNum=1'
    
    try:
        resp = requests.get(query_string)
        resp.raise_for_status()  # Raise an HTTPError for bad responses

        data_mrt = resp.json()

        if data_mrt['found'] > 0:
            mrt_lat.append(data_mrt["results"][0]["LATITUDE"])
            mrt_long.append(data_mrt["results"][0]["LONGITUDE"])
            print(f"{query_address}, Lat: {data_mrt['results'][0]['LATITUDE']} Long: {data_mrt['results'][0]['LONGITUDE']}")
        else:
            mrt_lat.append('NotFound')
            mrt_long.append('NotFound')
            print("No Results")

    except requests.RequestException as e:
        # Handle requests exceptions (including ConnectionError)
        mrt_lat.append('Error')
        mrt_long.append('Error')
        print(f"Request failed for {query_address}: {e}")

```

```python
# Create a DataFrame for MRT locations with their latitude and longitude
mrt_location = pd.DataFrame({
    'MRT': list_of_mrt,
    'latitude': mrt_lat,
    'longitude': mrt_long
})
```

```python
# Save MRT location data to a CSV file
mrt_location.to_csv('mrt_location.csv', index = False)
```

```python
address_list = combined_df['address'].unique() # Get unique addresses
```

```python
latitude = []
longitude = []
blk_no = []
road_name = []
postal_code = []
address = []
count = 0

# Loop through the address list and get coordinates and details for each address using the OneMap API
for row in range(len(address_list)):
    query_address = address_list[row]
    query_string='https://www.onemap.gov.sg/api/common/elastic/search?searchVal='+str(query_address)+'&returnGeom=Y&getAddrDetails=Y&pageNum=1'
    resp = requests.get(query_string)


    data_geo_location=json.loads(resp.content)
    if data_geo_location['found'] != 0:
        latitude.append(data_geo_location['results'][0]['LATITUDE'])
        longitude.append(data_geo_location['results'][0]['LONGITUDE'])
        blk_no.append(data_geo_location['results'][0]['BLK_NO'])
        road_name.append(data_geo_location['results'][0]['ROAD_NAME'])
        postal_code.append(data_geo_location['results'][0]['POSTAL'])
        address.append(query_address)
        print (str(query_address) + " ,Lat: " + data_geo_location['results'][0]['LATITUDE'] + " Long: " + data_geo_location['results'][0]['LONGITUDE'])
    else:
        print ("No Results")
```

```python
# Create a DataFrame with coordinates and details for each address
df_coordinates = pd.DataFrame({
    'latitude': latitude,
    'longitude': longitude,
    'blk_no': blk_no,
    'road_name': road_name,
    'postal_code': postal_code,
    'address': address
})
len(df_coordinates)
```

```python
# Get the latitude and longitude columns from the DataFrame for later distance calculation
list_of_lat = df_coordinates['latitude']
list_of_long = df_coordinates['longitude']
mrt_lat = mrt_location['latitude']
mrt_long = mrt_location['longitude']
```

```python
list_of_coordinates = []
list_of_mrt_coordinates = []
# Create tuples of (latitude, longitude) for each address and MRT station
for lat, long in zip(list_of_lat, list_of_long):
    list_of_coordinates.append((lat,long))
for lat, long in zip(mrt_lat, mrt_long):
    list_of_mrt_coordinates.append((lat, long))
```

```python
# Distance to nearest MRT station using geodesic distance (from the geopy library)
from geopy.distance import geodesic

list_of_dist_mrt = []
min_dist_mrt = []
# Calculate the distance from each address to every MRT station and store the shortest one
for origin in list_of_coordinates:
    for destination in range(0, len(list_of_mrt_coordinates)):
        list_of_dist_mrt.append(geodesic(origin,list_of_mrt_coordinates[destination]).meters)
    shortest = (min(list_of_dist_mrt))
    min_dist_mrt.append(shortest)
    list_of_dist_mrt.clear()
```

```python
# Distance from CDB
cbd_dist = []
# Use the coordinates of Singapore's CBD (1.2830, 103.8513) to calculate distance
for origin in list_of_coordinates:
    cbd_dist.append(geodesic(origin,(1.2830, 103.8513)).meters) #CBD coordinates
```

```python
# Add CBD and MRT distance columns to the coordinates DataFrame
df_coordinates['cbd_dist'] = cbd_dist
df_coordinates['min_dist_mrt'] = min_dist_mrt
```
